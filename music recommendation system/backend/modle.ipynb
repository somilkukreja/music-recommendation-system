```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
df['combined']=(
    df['song_name'].astype(str)+ ' ' + df['artist'].astype(str)+ ' ' + df['album'].astype(str)+ ' ' + df['genre'].astype(str)
)
```

```python
df['combined']
```

```python
tfidf=TfidfVectorizer(stop_words='english')
vectors=tfidf.fit_transform(df['combined'])
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommended(song):
    song = song.lower() # 4 spaces in
    
    try:
        # 8 spaces in (because it's inside 'try')
        index = df[df['song_name'].str.lower() == song].index[0]
    except:
        # 8 spaces in (because it's inside 'except')
        return []

    # These are all 4 spaces in
    distances = similarity[index]
    song_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    recommendation = []
    for i in song_list:
        # 8 spaces in (because it's inside 'for')
        recommendation.append(df.iloc[i[0]]['song_name'])
        
    return recommendation
```

```python
pickle.dump(df,open('songs_df.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
pickle.dump(tfidf,open('tfidf.pkl','wb'))

```
